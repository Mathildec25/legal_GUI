```python
import pandas as pd

# Chemin vers ton fichier brut
raw_path = "data/dataset_raw.csv"
clean_path = "data/dataset_clean.csv"

# ⚠️ adapte le séparateur ici si tu sais déjà lequel c'est
df = pd.read_csv(raw_path, sep="\t", encoding="utf-8-sig")

# Nettoyage colonnes : supprime espaces et virgules parasites
df.columns = [c.strip().strip(",;") for c in df.columns]

# Mapping pour renommer une bonne fois
rename_map = {
    "Name": "NAME",
    "Substance Name": "NAME",
    "CAS": "CAS",
    "CAS Number": "CAS",
    "IUPAC": "IUPAC",
    "IUPAC Name": "IUPAC",
    "Belgian annex": "Belgian Annex",
    "Belgian Annex": "Belgian Annex",
    "EU_Annex_Category_273_2004": "EU_Annex_Category_273_2004",
    "EU Annex Category 273/2004": "EU_Annex_Category_273_2004",
    "EU_Annex_Category_111_2005": "EU_Annex_Category_111_2005",
    "EU Annex Category 111/2005": "EU_Annex_Category_111_2005",
    "UNO": "UNO",
    "UN": "UNO",
    "Convention": "UNO",
    "SMILES": "SMILES",
    "Canonical_Smiles": "canonical_smiles",
    "Canonical SMILES": "canonical_smiles",
}

df = df.rename(columns=rename_map)

# Colonnes attendues par l'app (crée vide si manquante)
required = [
    "NAME","CAS","IUPAC","Belgian Annex",
    "EU_Annex_Category_273_2004","EU_Annex_Category_111_2005",
    "UNO","SMILES","canonical_smiles"
]
for col in required:
    if col not in df.columns:
        df[col] = pd.NA

# Sauvegarde propre
df.to_csv(clean_path, sep=",", index=False, encoding="utf-8-sig")

print("Fichier nettoyé enregistré :", clean_path)
print("Colonnes finales :", df.columns.tolist())
```
